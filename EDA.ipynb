## The distribution of other diseases that are comorbid with pneumonia

```python
comorbid_df = fixed_all_xray_df.loc[fixed_all_xray_df['Contains Pneumonia']==1]
```

```python
comorbid_df.sample(3)
```

```python
##Since there are many combinations of potential findings, I'm going to look at the 30 most common co-occurrences:
plt.figure(figsize=(16,7))
ax = sns.countplot(comorbid_df['Finding Labels'],
              order=comorbid_df['Finding Labels'].value_counts().iloc[:30].index)
plt.xticks(rotation=65,
           horizontalalignment='right',
           fontweight='light',
           fontsize='x-large'  )
plt.yticks(fontweight='light',
           fontsize='x-large')
for p in ax.patches: ax.annotate(np.round(p.get_height(),decimals=2), (p.get_x()+p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')
```

```python
comorbid_df['Finding Labels'].value_counts().iloc[:30]
```

```python
## Then I want to take a look about how these diseases influence each other.
all_labels = list(set(list(chain.from_iterable([x.split('|') for x in fixed_all_xray_df['Finding Labels']]))))
all_labels
```

```python
diagnoses_labels = all_labels.copy()
diagnoses_labels.remove('No Finding')
diagnoses_labels
```

```python
for label in all_labels:
    fixed_all_xray_df[label] = fixed_all_xray_df['Finding Labels'].map(lambda x : 1.0 if label in x else 0.0)
```

```python
fixed_all_xray_df.head()
```

```python
correlated_df = fixed_all_xray_df[diagnoses_labels].corr()
correlated_df
```

```python
plt.figure(figsize=(16,7))
sns.heatmap(correlated_df, cmap='Blues')
```

```python
## Show the most correlated Diseases with Pneumonia
correlated_df['Pneumonia'].nlargest(6)
```

```python
## Therefore, we can get the result that Pneumonia is highly correlated with Edema and Infiltration.
```

## Pixel-level assessments of the imaging data for healthy & disease states of interest (e.g. histograms of intensity values) and compare distributions across diseases.

```python
sample_df = pd.read_csv('sample_labels.csv')
all_image_paths = {os.path.basename(x): x for x in glob(os.path.join('/data','images*', '*', '*.png'))}
print('Scans found:', len(all_image_paths), ', Total Headers', sample_df.shape[0])
sample_df['path'] = sample_df['Image Index'].map(all_image_paths.get)
```

```python
sample_df.sample(3)
```

```python
## Take a few look for the first three raw imgs.
for ipath in sample_df.head(3)['path']:
    fig = plt.figure(figsize=(16,10))
    img = imread(ipath)
    ax2 = sns.distplot(img, bins=256)
    plt.show()
```

```python
all_labels_sample = list(set(chain.from_iterable([x.split('|') for x in sample_df['Finding Labels']])))
all_labels_sample
```

```python
diagnoses_labels_sample = all_labels_sample.copy()
diagnoses_labels_sample.remove('No Finding')
diagnoses_labels_sample
```

```python
for label in all_labels_sample:
    sample_df[label] = sample_df['Finding Labels'].map(lambda x: 1.0 if label in x else 0.0)
sample_df.head(10)
```

```python
sample_df.columns
```

```python
## Take a look in different diseases separetly.
## First, healthy testers
no_finding_sample = sample_df[sample_df['No Finding']==1.0]
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16,10))
fig.suptitle('Healthy Testers')
for ax, ipath in zip(axes.flatten(), no_finding_sample.head(6)['path']):
    img = imread(ipath)
    img = img[img>40]
    sns.distplot(img.ravel(), ax=ax, bins=256)
```

```python
diseases_patients = {}
for disease in diagnoses_labels_sample:
    diseases_patients[disease]=list(sample_df[sample_df[disease]==1.0]['path'])
```

```python
len(diagnoses_labels_sample)
```

```python
for i in diseases_patients:
    print(i+'\t'+diseases_patients.get(i)[0])
```

```python
## Group 1
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(16,19))
fig.suptitle("Group: 1")
for ax, ipath in zip(axes.flatten(), diseases_patients):
    color = ['red' if ipath == 'Pneumonia' else 'skyblue'][0]
    img = imread(diseases_patients.get(ipath)[0])
    img = img[img>40]
    sns.distplot(img.ravel(), ax=ax, bins=256, color=color)
    ax.set_title(ipath+'\n'+'mean='+str(round(img.ravel().mean(), 2))+', std='+str(round(img.ravel().std(), 2)))
```
